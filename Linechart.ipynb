```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('passing_cleaned.csv')

data = {
    'Year': [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Tom Brady': [526, 550, 601, 373, 602, 563, 620, 578, 0, 565, 492, 611, 637, 628, 619, 624, 432, 581, 570, 613, 610],
    'Peyton Manning': [571, 547, 591, 379, 497, 453, 557, 679, 0, 571, 679, 543, 583, 659, 597, 369, 0, 0, 0, 0, 0]
}
  
df = pd.DataFrame(data)

# Plot stacked area chart
plt.figure(figsize=(10, 6))
plt.stackplot(df['Year'], df['Tom Brady'], df['Peyton Manning'], labels=['Tom Brady', 'Peyton Manning'], colors=['red', 'blue'], alpha=0.8)

# Chart customization
plt.title("Cumulative Passing Attempts: Tom Brady vs. Peyton Manning")
plt.xlabel("Year")
plt.ylabel("Number of Passing Attempts")
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the chart
plt.show()
```
